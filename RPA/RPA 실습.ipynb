```python
os.listdir('./실습1/pdf_files/')
```

```python
# 실습 1
# - 파일명이 날짜와 시간으로 구성된 파일을 날짜별로 폴더를 만들어서 자동 분류하는 프로그램을 개발해주세요.
pdf_file_day = []
for i in range (1,len(os.listdir('./실습1/pdf_files/'))):
    pdf_file_day.append(os.listdir('./실습1/pdf_files/')[i].split('_')[0:4][1])
```

```python
pdf_file_day
```

```python
for i in range(0,len(pdf_file_day)):    
    file_name = os.listdir('./실습1/pdf_files/')[i+1]

print(file_name[i])
```

```python
for i in range(len(pdf_file_day)):
    os.makedirs(f'./실습1/pdf_files/{pdf_file_day[i]}',exist_ok=True)
```

```python
for i in range(len(pdf_file_day)):    
    file_name = os.listdir('./실습1/pdf_files/')[i+1]
    shutil.move(f'./실습1/pdf_files/{file_name[i]}', f'./실습1/pdf_files/{pdf_file_day[i]}')
```
